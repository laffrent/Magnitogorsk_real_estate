```python
df = df.dropna(how='all')
```

```python
df['apartment'] = df['apartment'].fillna('unknown')
df['note'] = df['note'].fillna('no_note')

df['rooms'] = df['apartment'].apply(lambda x: x.split()[0])
```

```python
def fill_appart(string):
    
    string = string.lower()
    if 'студия' in string:
        return 'Однокомнатная'
    elif '2-х' in string:
        return 'Двухкомнатная'
    elif 'пятикомн' in string:
        return 'Пятикомнатная'
    else:
        return 'unknown'
    # elif 'дач' in string or 'дач' in string:
    #     return 'Пятикомнатная'
```

```python
df.loc[df['rooms'] == 'unknown', 'rooms'] = df.loc[df['rooms'] == 'unknown', 'note'].apply(fill_appart)
#df.loc[df['rooms'] == 'Многокомнатная', 'rooms'] = df.loc[df['rooms'] == 'unknown', 'note'].apply(fill_appart)
```
